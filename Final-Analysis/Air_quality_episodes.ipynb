```python

fig, ax = plt.subplots(figsize = (10,10))
plt.scatter([0,1,2], [2.4299,0.5462,0.469],color='red', marker = 'd', label = 'O$_3$', s =100 )
plt.scatter([0,1,2], [0.15267175572519084,1.9115044247787611,2.22], color = 'k', marker = 'o', label ='PM$_{2.5}$', s =100)
plt.title('Relative Likelihood Of Air Quality Episodes - Urban Winter', fontsize =23, y =1.1)
plt.ylim([0,3.5])
plt.hlines([1,2,3],xmin = -0.5, xmax =3.2,linestyles = 'dashed', color = 'k')
plt.xlim([-.5,2.5])
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
plt.legend(prop={'size': 12})
ax.set_xticks([0,1,2])
ax.set_yticks([1,2,3])
ax.set_ylabel('Relative Likelihood of Air Quality Episode', fontsize =19)
ax.set_yticklabels(labels = [r'$1.0\times $',r'$2.0\times$',r"$3.0\times$"], fontsize = 16)
ax.set_xticklabels(labels = ['NAO+','NAO-',r"Blocking"], fontsize = 16)
plt.savefig('Likelihood_air_qual_episode.png', dpi=500)
plt.show()
```
